```python
"""
    File : heatmapFitness.ipynb
    Plot : heatmap of average fitness of all tumour voxels in a tumour section among replicate simulations
    Inputs : (related to several figure panels)
        SourceData_Fig2c_part2.xlsx, 
        SourceData_ExtendedData_Fig1c_part2.xlsx
        SourceData_ExtendedData_Fig4a_part2.xlsx
        SourceData_ExtendedData_Fig4b_part2.xlsx
"""
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import os
```

```python
path_to_all_source_data = "../Source data for figures"
```

```python
def read_and_plot(sheet_names):
    for sheet_name in excelfile.sheet_names:
        data = pd.read_excel(
            excelfile, 
            sheet_name=sheet_name,
            header=None
        )

        print(sheet_name)

        fig = plt.figure(dpi=300)
        ax = fig.add_subplot(111)
        ax.set_aspect('equal')
        ax.imshow(data, cmap=plt.cm.Blues, vmin=0.25, vmax=1)
        plt.axis('off')
        plt.show()
        plt.close()
```

# Figure 2c (fitness heatmap)

```python
path_to_excelfile = os.path.join(
    path_to_all_source_data,
    "Source_Data_Figure_2",
    "SourceData_Fig2c_part2.xlsx"
)
```
