```python
data_fig2bi = pd.read_excel(
    excelfile, 
    sheet_name='SourceDataFig2bi',
    header=None
)
data_fig2bii = pd.read_excel(
    excelfile, 
    sheet_name='SourceDataFig2bii',
    header=None
)
```

### plotting

```python
def config_font_size(figsize):
    params = {'legend.fontsize': 6,
              'figure.figsize': figsize,
             'axes.labelsize': 8,
             'axes.titlesize': 8,
             'xtick.labelsize':6,
             'ytick.labelsize':6,
             'pdf.fonttype':42}
    plt.rcParams.update(params)
    
def config_ticks(ax):
    # xtick labels
    xticklabels = [
        "0e-4", "1e-4", "2e-4", 
        "3e-4", "4e-4", "5e-4", 
        "6e-4", "7e-4", "8e-4", 
        "9e-4", "1e-3"
    ]
    # ytick labels
    yticklabels = [
        '0', '0.2', '0.4', '0.6', '0.8', '1'
    ]
    
    ax.set_yticks(range(len(yticklabels)))
    ax.set_xticks(range(len(xticklabels)))
    ax.set_yticklabels(yticklabels)
    ax.set_xticklabels(xticklabels, rotation = 45)
    
    ax.set_xlabel("Driver acquisition probability (p_driver)")
    ax.set_ylabel("Selective coefficient (s)")
```

```python
fig = plt.figure(dpi=300)
ax = fig.add_axes([0.2,0.3,0.75,0.65])
cax = fig.add_axes([0.3,0.15,0.55,0.05])

config_font_size(figsize = (2.75,3))
config_ticks(ax)

im = ax.imshow(
    data_fig2bi,
    cmap = plt.cm.Greens
)

cbar = plt.colorbar(im, cax=cax, orientation='horizontal')
cbar.ax.set_xlabel("Number of subclones")

plt.show()
```
