```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os
import seaborn as sns
```

```python
path_to_all_source_data = "../Source data for figures"
```

```python
def plot_data(data, ref_shannon, xticklabels, figsize, ylim=None):
    def config_font_size(figsize):
        params = {'legend.fontsize': 6,
                  'figure.figsize': figsize,
                 'axes.labelsize': 8,
                 'axes.titlesize': 8,
                 'xtick.labelsize':6,
                 'ytick.labelsize':6,
                 'pdf.fonttype':42}
        plt.rcParams.update(params)

    fig = plt.figure(dpi=300)
    ax = fig.add_axes([0.1,0.2,0.85,0.75])

    config_font_size(figsize = figsize)

    ax.axhline(y=ref_shannon, ls="--", c='grey', lw = 0.5)

    sns.boxplot(ax = ax, data = data, x = "Label", y = "Shannon")
    sns.swarmplot(ax = ax, data = data, x = "Label", y = "Shannon", size = 1.5, color="white")

    ax.set_xlabel('Starting tumour diameter for re-growth (mm)')
    ax.set_ylabel('Shannon diversity')

    ax.set_xticklabels(xticklabels)
    if ylim:
        ax.set_ylim(ylim)

    plt.show()
```

# Extended Data Figure 10

```python
path_to_excelfile = os.path.join(
    path_to_all_source_data,
    "Source_Data_Extended_Data_Figure_10",
    "SourceData_ExtendedData_Fig10.xlsx"
)
```

## ED Fig10d

```python
data = pd.read_excel(excelfile, sheet_name='SourceDataEDFig10d')
reference_shannon = 2.384931554358255
evolution_replay_starting_sizes = [
    '5.41', '11.34', '16.43', '22.37', '28.21', '34.60', '40.54', '47.15', '54.35', '62.74', '72.19', '83.74', '96.59', '110.05'
]
```

```python
plot_data(
    data=data, 
    ref_shannon=reference_shannon, 
    xticklabels=evolution_replay_starting_sizes,
    ylim=(0,3),
    figsize=(8.5,3)
)
```
